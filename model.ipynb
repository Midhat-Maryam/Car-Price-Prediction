```python
df['name']=df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df=df.reset_index(drop=True)
```

```python
df['year'].unique()
```

```python
df=df[df['year'].astype(str).str.isnumeric()]
```

```python
df['year']=df['year'].astype(int)
```

```python
df['kms_driven'].unique()
```

```python
df['kms_driven']=df['kms_driven'].str.split(" ").str.get(0).str.replace(",",'')
```

```python
df = df.dropna(subset=['kms_driven'])
```

```python
df=df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven']=df['kms_driven'].astype(int)
```

```python
df['fuel_type'].unique()
```

```python
df=df[ df['fuel_type'].notna()]
```

```python
df['company'].unique()
```

```python
df['Price'].unique()
```

```python
df=df[df['Price']!='Ask For Price']
```

```python
df['Price'] = df['Price'].str.replace(",","").astype(int)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
num_cols=['year','Price','kms_driven']
```

```python
for col in num_cols:
    plt.figure(figsize=(12, 6))  
    plt.hist(df[col])
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(12, 6))  
    plt.boxplot(df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)][col]

    if len(outliers) > 0:
        print(f"Column: {col}")
        print(f"   Outliers detected: {len(outliers)}")
        print(f"   Example outliers: {outliers.head().tolist()}")
        print("-" * 50)
```

```python
df=df[df['Price']<6000000]
```

```python
df.to_csv('Cleaned Car.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
categorical_cols = ['company','fuel_type']
numeric_cols = ['year','kms_driven']
```

```python
transformer=ColumnTransformer([
    ("One Hot Encoding",OneHotEncoder(sparse_output=False,drop="first",handle_unknown="ignore"),categorical_cols),
    ("scaler", StandardScaler(), numeric_cols)
])
```

```python
pipe = Pipeline([
    ("transformer", transformer),
    ("model", LinearRegression())
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
```
